```python
from langchain_community.document_loaders import WebBaseLoader
import requests
from bs4 import BeautifulSoup
```

```python
links=[]
```

```python
with open(file=r"D:\Lang_chain\Healthcare_Assistant_Chatbot\data_files\links.txt",mode="r") as f:
    for i in f:
        if i.strip()!='':
            links.append((i.split('-'))[1].strip())
```

```python
links
```

```python
loader=WebBaseLoader(web_path=links[0])
loaded=loader.load()
loaded
```

```python
final_text_data=[]
```

```python
def data_extraction(url):

    response=requests.get(url)
    soup=BeautifulSoup(response.text,"html.parser")

    for h2 in soup.find_all("h2"):
        if "References" in h2.get_text():
            for sibling in h2.find_next_siblings():
                sibling.decompose()
            h2.decompose()
            break 

    for sup in soup.find_all("sup", class_="reference"):
        sup.decompose()

    content = soup.find_all("p")
    clean_text = "\n".join(p.get_text() for p in content)
    
    return clean_text
```

```python
for i in links:
    dat=data_extraction(i)
    final_text_data.append(dat)
```
