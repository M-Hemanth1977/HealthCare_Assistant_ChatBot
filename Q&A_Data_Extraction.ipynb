```python
import json

with open(r"D:\Lang_chain\Healthcare_Assistant_Chatbot\data_files\data.json",mode="r") as file:
    info=json.load(file)
```

```python
data=json.dumps(info,indent=1)
data
```

```python
import import_ipynb
from web_data_extraction import complete_data
```

```python
complete_data
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter=RecursiveCharacterTextSplitter(separators=["\n\n","\n"],chunk_size=1000,chunk_overlap=200)

splitted=splitter.split_text(text=data)
```

```python
splitted
```

```python
len(splitted)
```

```python
splitted_1=splitter.split_text(complete_data)
splitted_1
```

```python
combined_splitted_lists=splitted+splitted_1
```

```python
len(combined_splitted_lists)
```

```python
from langchain_groq import ChatGroq

llm=ChatGroq(model="llama3-8b-8192",api_key=os.getenv("groq_api_key"))

llm
```

```python
from langchain_core.prompts import ChatPromptTemplate

template=ChatPromptTemplate.from_messages(
    [
        ("system","""
         
         You are an HealthCare AI Assitant expert and good at answering questions related to healtcare related questions.
         Based on the input text provided you need to generate 2 Question and Answer pairs 
         and show only question and correspnsing answer and also make sure that each question
         should be specific and relevant to the content and give in some brief like 2 to 3 lines.
         Focus on:
         - symptoms and diagnosis
         - causes and risk factors
         - treatments or medications etc...
         The Output Format is shown as below:
         Q1: ...
         A1: ...
         Q2: ...
         A2: ...
         Q3: ...
         A3: ...   
                                            """),
        ("user","{input}")
    ]
)

template
```

```python
from langchain_core.output_parsers import StrOutputParser

parser=StrOutputParser()

parser
```

```python
chain=template|llm|parser
chain
```

```python
chain.input_schema
```

```python
all_outputs=[]

for index,sentence in enumerate(combined_splitted_lists):
    
    print(f"Processing Batch-{index+1}/{len(combined_splitted_lists)}...")

    try:
        response=chain.invoke({"input":sentence})
        all_outputs.append(response)
        print(f"Batch-{index+1} processing done.")

    except Exception as e:
        print(f"Batch-{index+1} processing failed")
```

```python
with open("Q&A outputs.txt","w",encoding="utf-8") as f:
    f.write("\n".join(all_outputs))
```
